# MSE formula

- ### Step-1 : we already has y_actual (y_test) data and y_prediction  (model predictions)
- ### Step-2 : throungh a for loop will call each value and subtract both 
- ### Step-3 : square the subtractions and append in a list 
- ### Step-4 : get the sum of all the values in a list
- ### Step-5 : Divide by len(y_test) or len(y_predictions )

```python
s=0
for i in range(len(y_test)):
    v1=y_test.values[i]-y_predictions[i]
    v2=v1**2
    s=s+v2
print(s/len(y_test))
```

```python
#step-1: import package
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# step-2 :
df1="C:\\Users\\admin\\OneDrive\\PRATIK\\NARESH IT\\DATA SCIENCE\\DATA SETS\\csv\\winequality_red.csv"
vine_df=pd.read_csv(df1)
vine_df

# step-:3
vine_df.dropna(inplace=True)
vine_df.reset_index(inplace=True)
vine_df.drop('index',axis = 1,inplace=True)

# step- :4
X=vine_df.drop('quality',axis=1)
y=vine_df['quality']

# step 4  :
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,   # input data
                                               y,   # output data
                                               random_state=1234, # it select 
                                               test_size=0.30)

# step 6 : develope the model
from sklearn.linear_model import LinearRegression
LR=LinearRegression()
LR.fit(X_train,y_train)

# step 7 : save the model
import pickle
pickle.dump (LR,
            open ('linear_model.pkl','wb'))

# step 8 : load
model = pickle.load(open('linear_model.pkl', 'rb'))
model
```
